```python
file=open("a.dat")
i=0
while True:
    if i==7:
        f=file.readline()
        break
    file.readline()
    i+=1
arr=f.split(" ")
ax1=[]
ay1=[]
c=int(len(arr)/2)
for i in range(c):
    ax1.append(float(arr[i]))
for i in range(c,len(arr)):
    ay1.append(float(arr[i]))
plt.scatter(ax1,ay1)
plt.plot(ax1,ay1)
```
